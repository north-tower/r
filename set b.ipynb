```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Download the data as a csv file
data= pd.read_csv("C:/Users/Miki/Downloads/SET_B.csv")
```

```python
#Checking if the data could be used for linear regression
years = data['Year']
age = data['Female']
plt.scatter(years,age,label="Age")
plt.title("Year vs Life expectancy")
```

```python
#Drawing the line for linear regression
slope, intercept, r, p, std_err = stats.linregress(years, age)
def myfunc(years):
  return slope * years + intercept

mymodel = list(map(myfunc, years))

plt.scatter(years, age)
plt.plot(years, mymodel)
plt.show()
```

```python
#Getting the realationship between the x-axis and the y-axis
print(r)
```

```python
#predict the life expectancy for females in 2030
age = myfunc(2030)

print(age)
```
