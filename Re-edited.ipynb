```python
import pandas as pd
```

```python
#importing the dataset
df = pd.read_excel("Readmissions2021.xlsx")
df.head(10)
```

```python
#30 day rolling number of readmissions
df['readmission'] = df.set_index('admission_date').sort_index()\
                .groupby('patient_id', sort=False)['patient_id']\
                .rolling('30d', closed='both').count().tolist()
#Removing the intial admission
df.loc[:, 'readmission'] = df['readmission'].apply(lambda x: x - 1)


df[['patient_id','readmission']].head(10)
```

```python
#Binary Readmission
df.loc[df['readmission'] > 0, 'readmission'] = 1

df.loc[df['readmission'] == 0, 'readmission'] = 0


df[['patient_id','readmission']].head(25)
```
