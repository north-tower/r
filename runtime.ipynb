```python
import matplotlib.pyplot as plt
import numpy as np
import datetime
import time
import timeit
import random
import statistics
```

```python
#Function one
def f1(lst):
    r = 0;
    n = 100;
    if len(lst) < 100:
        n = len(lst);
    for x in range(n):
        r += x
```

The complexity of the function f1 is O(c) where c is a constant. So the function has constant running time

```python
#Gathering the timings
ns = []
ts = []
lst = []

# calculating the execution time for the input 100
for i in range(1000):

    n = random.randint(1,200)

    lst = random.sample(range(1,1000),n)

    start = time.time()

    f1(lst)

    end = time.time()

    ns.append(n)

    ts.append(end-start)
```

```python
#Plot your data and bounding functions here
plt.plot(ns, ts, 'or')
plt.plot(ns, [0.001 * n for n in ns], '-g')
plt.plot(ns, [0.005 * n for n in ns], '-b')
```

```python
#Function two
def f2(x):
 r = x / 2
 d = 1e-10
 while abs(x - r**2) > d:
     r = (r + x/r) / 2
 return r
```

Hypothesis: The worst case complexity is O(n) as worst case complexity.

```python
#Function Three
def f3(lst):
 while True:
     swapped = False
     for i in range(len(lst)-1):
         if lst[i] > lst[i+1]:
             lst[i], lst[i+1] = lst[i+1], lst[i]
             swapped = True
     if not swapped:
         break
```

Hypothesis:O(n) is taken as the worst case complexity in this function

```python
testing_array = [0]
ns=[]
ts=[]
for i in range(1, 120):
    start = time.time()
    f3(testing_array)
    end = time.time()
    ts.append(end - start)
    ns.append(len(testing_array))
    testing_array.append(i)
```

```python
# plot your data and bounding functions here
plt.plot(ns, ts, 'or')
plt.plot(ns, [0.08 * n for n in ns], '-g')
plt.plot(ns, [0.12 * n for n in ns], '-b');
```

```python
#Function Four
def f4(n):
     counters = [0] * n
     while not all(counters):
         for i in range(n):
             if counters[i]: 
                 counters[i] = 0
             else:
                 counters[i] = 1
                 break
```

Hypothesis:f= O(n) (as worst case complexity)

```python
ns = np.linspace(10, 10_000, 50, dtype=int)
ts = [timeit.timeit('f4({})'.format(n), number=100,globals=globals() )for n in ns]

plt.plot(ns, ts, 'or')
plt.plot(ns, [0.0006 * n for n in ns], '-g')
plt.plot(ns, [0.0009 * n for n in ns], '-b')
```
